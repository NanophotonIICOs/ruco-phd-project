```python
def rdat(pathdata):
    try:
        Exp          = np.loadtxt(pathdata,delimiter=',')
    except :
        Exp          = np.loadtxt(pathdata,delimiter='\t')
    return Exp[:,:]

def rxlsdata(pathdata):
    Exp   =  pd.read_excel(pathdata,index=False)
    Expn  =  Exp.values
    return Expn
```

```python
def refractive_index(name,T,xc,l):
   # Global parameters
   eVtoMic = 1/1.239856
   EG0     = 1.5192*eVtoMic
   EDeb    = 0.0159*eVtoMic
   ETO     = 0.0336*eVtoMic
   S       = 1.8
   STO     = 1.1
   KB      = 0.0000861708*eVtoMic
   A0T     = 5.9613 + (0.0007178)*T - (0.000000953)*(T**2)
   E102T   = 4.7171 - (0.0003237)*T - (0.000001358)*(T**2)
   #EGGaAsT = EG0 + S*EDeb*((1-np.cosh(EDeb/2*KB*T))/np.sinh(EDeb/2*KB*T)) + STO*ETO*((1-np.cosh(ETO/2*KB*T))/np.sinh(ETO/2*KB*T))
   EGGaAsT = EG0 + S*EDeb*(1-coth(EDeb/2*KB*T))+STO*ETO*(1-coth(ETO/2*KB*T))
   #Dependence of composition x of Al_xGa _ (1 - x) As
   E22  = 0.000724
   C2   = 0.00155
   E32  = 0.001331
   C3   = 0.00261
   if  name == 'AlGaAs':
       x     = xc
       AXT   = A0T-16.159*x + 43.511*x**2 - 71.317*x**3 + 57.535*x**4 - 17.451*x**5
       C1T   = 21.5647 + 113.74*x - 122.5*x**2 + 108.401*x**3 - 47.3*x**4
       E12T  = E102T + 11.006*x - 3.08*x**2
       C0T   = 1/(50.535 - 150.7*x - 62.209*x**2 + 797.16*x**3 - 1125*x**4 + 503.79*x**5)
       E0T2  = (EGGaAsT + 1.1308*x + 0.1436*x**2)**2
       R     = (1-x)*C2/(E22-(1000/l)**2) + x*C3/(E32-(1000/l)**2)
       nalgaas = (-4.0728835e-5* l+1.03)*1.004*cmath.sqrt(AXT+(C0T/(E0T2-(1000/l)**2))+(C1T/(E12T-(1000/l)**2))+R)
       return nalgaas
   elif name == 'GaAs':
       x     = 0
       AXT   = A0T-16.159*x + 43.511*x**2 - 71.317*x**3 + 57.535*x**4 - 17.451*x**5
       C1T   = 21.5647 + 113.74*x - 122.5*x**2 + 108.401*x**3 - 47.3*x**4
       E12T  = E102T + 11.006*x - 3.08*x**2
       C0T   = 1/(50.535 - 150.7*x - 62.209*x**2 + 797.16*x**3 - 1125*x**4 + 503.79*x**5)
       E0T2  = (EGGaAsT + 1.1308*x + 0.1436*x**2)**2
       R     = (1-x)*C2/(E22-(1000/l)**2) + x*C3/(E32-(1000/l)**2)
       ngaas = (-4.0728835e-5*l+1.03)*1.004*cmath.sqrt(AXT+(C0T/(E0T2-(1000/l)**2))+(C1T/(E12T-(1000/l)**2))+R)
       return ngaas
   elif name == 'AlAs':
       x     = 1
       AXT   = A0T-16.159*x + 43.511*x**2 - 71.317*x**3 + 57.535*x**4 - 17.451*x**5
       C1T   = 21.5647 + 113.74*x - 122.5*x**2 + 108.401*x**3 - 47.3*x**4
       E12T  = E102T + 11.006*x - 3.08*x**2
       C0T   = 1.0/(50.535 - 150.7*x - 62.209*x**2 + 797.16*x**3 - 1125*x**4 + 503.79*x**5)
       E0T2  = (EGGaAsT + 1.1308*x + 0.1436*x**2)**2
       R     = ((1-x)*C2)/(E22-(1000/l)**2) + (x*C3)/(E32-(1000/l)**2)
       nalas = ((-4.0728835e-5)*l+1.03)*1.004*cmath.sqrt(AXT+(C0T/(E0T2-(1000/l)**2))+(C1T/(E12T-(1000/l)**2))+R)
       return nalas
```

```python
def structure(l):
   subs   = [200,   refractive_index('GaAs',30,0,l)]
   Cap1   = [600,   refractive_index('AlGaAs',30,0.15,l)]
   Cap2   = [300,   refractive_index('AlGaAs',30,0.15,l)]
   qw1    = [11.87, refractive_index('GaAs',30,0,l)]
   barr   = [1.98,  refractive_index('AlGaAs',30,0.15,l)]
   qw2    = [13.85, refractive_index('GaAs',30,0,l)]
   Cap3   = [300,   refractive_index('AlGaAs',30,0.15,l)]
   Cap4   = [10,    refractive_index('GaAs',30,0,l)]
   air    = [10,1]
   struct = [air,Cap4,Cap3,qw2,barr,qw1,Cap2,Cap1,subs]
   return np.array(struct)

nolay     = len(structure(680))
```

```python
def Dl(ni,theta,pol):
    if theta > 0:
       Di = np.empty((2,2),dtype = complex)
       if  pol == 's':
           Di[0,0]  =  1
           Di[0,1]  =  1
           Di[1,0]  =  ni*np.cos(theta)
           Di[1,1]  = -ni*np.cos(theta)
           return np.asmatrix(Di,dtype = complex)
       elif pol == 'p':
           Di[0,0]  =  np.cos(theta)
           Di[0,1]  =  np.cos(theta)
           Di[1,0]  =  ni
           Di[1,1]  = -ni    
           return np.asmatrix(Di,dtype = complex)
    else:
        Di = np.empty((2,2),dtype = complex)
        Di[0,0]  =  1
        Di[0,1]  =  1
        Di[1,0]  =  ni
        Di[1,1]  = -ni
        return np.asmatrix(Di,dtype = complex)
def Pl(ni,dl,l,theta):
   Pm       = np.empty((2,2),dtype = complex)
   kl       = 2*ni*np.pi*np.cos(theta)/l
   psi      = kl*dl
   Pm[0,0]  = np.exp(1j*psi)
   Pm[0,1]  = 0
   Pm[1,0]  = 0
   Pm[1,1]  = np.exp(-1j*psi)
   return  np.asmatrix(Pm,dtype = complex)
```

```python
# Parameters
lini     = 500
lfin     = 900
ran      = lfin-lini
step     = 0.1
Nostps   = ran/step
Rs       = np.zeros((int(Nostps),2),dtype = np.float32)
Rp       = np.zeros((int(Nostps),2),dtype = np.float32)
th       = 0
D0s      = Dl(1,th,'s')
D0p      = Dl(1,th,'p')
Ms       = inv(D0s)
Mp       = inv(D0p)
Ms       = np.asmatrix(Ms)
Mp       = np.asmatrix(Mp)
lvec     = np.arange(lini,lfin,step,dtype= np.float32) 



for ii in tqdm(range(len(lvec))):
   Ms   = inv(D0s)
   Mp   = inv(D0p)
   li   = lvec[ii]
   lays = structure(li)
   for jj in range(1,nolay-1):
       ni     = lays[jj,1]
       di     = lays[jj,0]
       thi    = sp.arcsin(1*np.sin(th)/ni)
       Pli    = Pl(ni,di,li,thi)
       Dsi    = Dl(ni,thi,'s')
       IDsi   = inv(Dl(ni,thi,'s'))
       Dpi    = Dl(ni,thi,'p')
       IDpi   = inv(Dl(ni,thi,'p'))
       Ms     = Ms*Dsi*Pli*IDsi
       Mp     = Mp*Dpi*Pli*IDpi
   nii      = lays[-1,1]
   thii     = sp.arcsin(ni*np.sin(th)/nii)
   Dsubs    = Dl(nii,thii,'s')
   Dsubp    = Dl(nii,thii,'p')
   Ms       = Ms*Dsubs 
   Mp       = Mp*Dsubp
   rsi      = Ms[1,0]/Ms[0,0]
   rpi      = Mp[1,0]/Mp[0,0]
   Rs[ii,0]  = lvec[ii]
   Rs[ii,1]  = abs(rsi)**2
   Rp[ii,0]  = lvec[ii]
   Rp[ii,1]  = abs(rpi)**2
 
```

```python
fig,ax1 = plt.subplots(1,1,figsize=(10,5))
ax1.plot(Rp[:,0],Rp[:,1],'b',label= '$abs$')

ax1.legend()
plt.show()
```
